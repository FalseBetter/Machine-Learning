## 기존 전처리한 .ipynb파일이 있었으나 현재 파일이 손상되어 안 열림..
- 다행이 csv파일은 보조되어 다시 전처리
- 기존 csv파일에서 형법범 데이터만 가져옴 나머지는 제외시킴(특별법범)
- 해당 csv파일은 2022년도

```python
import pandas as pd

df = pd.read_csv('지역별 범죄발생 data.csv', encoding='euc-kr')
df.head(3)
```

```python
df.columns
```

```python
# "지역별"이라는 상위 레벨을 추가할 컬럼들
data = ['서울', '부산', '대구', '인천', '광주', '대전', '울산', 
          '세종', '경기', '강원', '충북', '충남', '전북', '전남', 
          '경북', '경남', '제주']

# 멀티레벨 컬럼 인덱스 생성
new_columns = []

for col in df.columns:
    if col in data:
        new_columns.append(('지역별', col))  # 지역별이라는 상위 레벨 추가
    else:
        new_columns.append(('', col))  # 기타 컬럼은 상위 레벨 없이 유지

# 새로운 멀티인덱스 설정(가로형태 상위 컬럼)
df.columns = pd.MultiIndex.from_tuples(new_columns)
df.head()
```

## 지역별 범죄건수가 가장 많은 지역 조회

```python
# 지역별 데이터에서 각 도시별 범죄건수 합 (위에서 멀티인덱스로 지역별 컬럼에 지역포함)
df_total = df['지역별'].sum(axis=0) 
# 계산된 합계를 DataFrame으로 변환
df_total = df_total.to_frame(name='범죄건수')
# 범죄건수를 기준으로 내림차순
df_total_sort_values = df_total.sort_values(by='범죄건수', ascending=False)
df_total_sort_values
```
