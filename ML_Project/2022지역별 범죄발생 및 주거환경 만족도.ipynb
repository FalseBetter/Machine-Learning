## 2022거주_지역의_주거환경_만족도

```python
import pandas as pd

df = pd.read_csv('2022주거환경_만족도.csv', encoding='euc-kr')
df
```

```python
# data type 확인
df.info(), df.columns
```

```python
# 데이터의 컬럼 이름의 앞뒤 공백 제거
df.columns = df.columns.str.strip()

# 불필요한 컬럼 제거, 변경
df.drop(['특성별'], axis=1, inplace=True)

df_area = ['수도권', '비수도권', '서울', '부산', '대구', '인천', '광주', '대전', '울산', 
          '세종', '경기', '강원', '충북', '충남', '전북', '전남', 
          '경북', '경남', '제주']

# Unnamed: 1 컬럼에 지역명 제외하고 가져옴
df_filter = df[df['Unnamed: 1'].isin(df_area)]

# Unnamed: 1 컬럼명 변경 -> 지역별
df_filter.rename(columns={'Unnamed: 1' : '지역별'}, inplace=True)

# 인덱스 리셋
df_filter = df_filter.reset_index()
# 'index' 컬럼 삭제
df_filter.drop('index', axis=1, inplace=True)
df_filter
```

```python
# 지역별 치안 및 범죄 등 방범 상태 만족도가 높은 순으로 내림차순
df_filter.sort_values(by=['치안 및 범죄 등 방범 상태'], ascending=False)
```

## 두 데이터프레임 merge 하기전 같은 방향으로 변환

```python
df = pd.read_csv('[수정 전]2022지역별 범죄발생 data.csv', encoding='euc-kr')
df.head()
```

```python
# 2022지역별 범죄발생 data를 세로로 돌리기
data = ['서울', '부산', '대구', '인천', '광주', '대전', '울산', '세종', '경기', '강원', '충북', '충남', '전북', '전남', '경북', '경남', '제주']
df_crime = df.melt(id_vars=[col for col in df.columns if col not in data], var_name='지역명')

# '지역명' 컬럼을 추출하고 원래 데이터프레임에서 제거
df_area = df_crime.pop('지역명')

# '지역명' 컬럼을 데이터프레임의 맨 오른쪽으로 추가
df_crime.insert(0, '지역명', df_area)
df_crime.head(10)
```

```python
# 데이터의 컬럼 이름의 앞뒤 공백 제거
df_crime.columns = df_crime.columns.str.strip()
```

```python
df_crime.columns
```

```python
# 'value' 컬럼 삭제
df_crime.drop(['value'], axis=1, inplace=True)
# Unnamed: 0 컬럼 삭제
df_crime.drop(['Unnamed: 0'], axis=1, inplace=True)
# 범죄 대분류 삭제(생각해보니 굳이 필요 없을 거 같음)
df_crime.drop(['범죄 대분류'], axis=1, inplace=True)
# 지역명 컬럼 지역별 컬럼으로 변환
df_crime.rename(columns={'지역명' : '지역별'}, inplace=True)
# 합계 컬럼을 건수 컬럼으로 변환
df_crime.rename(columns={'합계' : '범죄건수'}, inplace=True)
df_crime
```

```python
# 저장
df_crime.to_csv('2022지역별_범죄발생 data.csv', encoding='euc-kr')
```

## 데이터 프레임 merge

```python
# 두 데이터프레임을 '지역별' 컬럼을 기준으로 합치기
df = pd.merge(df_filter, df_crime, on='지역별', how='inner')
df.head()
```

```python
# 컬럼 리스트를 생성
item = list(df.columns)

# '지역별' 컬럼의 인덱스를 찾음
index = item.index('지역별')

# '범죄 유형', '죄명', '합계' 컬럼을 지역별 오른쪽으로 이동
# 지역별 바로 다음 위치로 이동시키기 위해 index+1 위치로 각 컬럼을 삽입
for col in ['범죄 유형', '죄명', '범죄건수']:
    item.remove(col)  # 먼저 해당 컬럼을 리스트에서 제거
    item.insert(index + 1, col)  # 지역별 다음 위치에 삽입
    index += 1  # 다음 컬럼을 위해 인덱스 증가

# 데이터프레임의 컬럼 순서를 조정한 리스트로 재배열
df = df[item]
df.head()
```

```python
# 합친 데이터 프레임 저장
df.to_csv('2022지역별 범죄발생 및 주거환경 만족도.csv', encoding='euc-kr')
```
